```python
# Date: 2021-08-08 
# Favorite Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import decimal
```

```python
#LOOK AT READ ME FOR DEFNINITIONS 
all_seasons = pd.read_csv('all_seasons.csv')
all_seasons.head(5)
```

```python
print(all_seasons.isnull().sum())
print("Number of Duplicates",all_seasons.duplicated().sum())
all_seasons.dtypes
```

```python
all_seasons['age'] = all_seasons['age'].astype(np.int64)
all_seasons.dtypes
```

```python
#Older players have better stats vs younger ones.
#What college produced best players
#Draft Round vs performance
#Top 10 Americans vs Top 10 Non Americans
```

```python
#Age
maxage = (all_seasons['age'].max())
minage = (all_seasons['age'].min())
#Points
maxpts = (all_seasons['pts'].max())
minpts = (all_seasons['pts'].min())
#Rebounds
maxreb = (all_seasons['reb'].max())
minreb = (all_seasons['reb'].min())
#Assits
maxast = (all_seasons['ast'].max())
minast = (all_seasons['ast'].min())
```

```python
#Players age in correlation to their performance.
fig = plt.figure(figsize=(15,15)) 
fig_dims = (2, 2)
plt.suptitle('Age relation to performance', fontsize=20)

graph1 = plt.subplot2grid(fig_dims, (0, 0))

plt.scatter(all_seasons['age'],  all_seasons['pts'])
graph1.set_title('Points')
graph1.set_xticks(np.arange(minage,maxage,4))
graph1.set_yticks(np.arange(minpts,maxpts,2))
graph1.set_xlabel('Age')
graph1.set_ylabel('Points')

graph2 = plt.subplot2grid(fig_dims, (0, 1))

plt.scatter(all_seasons['age'],  all_seasons['reb'])
graph2.set_title('Rebounds')
graph2.set_xticks(np.arange(minage,maxage,4))
graph2.set_yticks(np.arange(minreb,maxreb,1))
graph2.set_xlabel('Age')
graph2.set_ylabel('Rebounds')

graph3 = plt.subplot2grid(fig_dims, (1, 0))

plt.scatter(all_seasons['age'],  all_seasons['ast'])
graph3.set_title('Assits')
graph3.set_xticks(np.arange(minage,maxage,4))
graph3.set_yticks(np.arange(minast,maxast,1))
graph3.set_xlabel('Age')
graph3.set_ylabel('Assits')
plt.show()
```

```python
#What college produced best players
all_seasons['OverallScore'] = all_seasons['pts'] 
+ all_seasons['reb']+ all_seasons['ast']
```

```python
all_seasons.head(5)
```

```python
# Get indexes where name column has value None
indexName = all_seasons[all_seasons['college'] == 'None'].index

# Delete these row indexes from dataFrame
all_seasons.drop(indexName , inplace=True)

test = all_seasons[['college','OverallScore']].groupby(['college'])['OverallScore'].sum().nlargest(10).reset_index(name = 'Overall')

print(test
```

```python
#These are the colleges that produce players withthe best overall score between points,rebounds,assits
plt.plot(test['college'],test['Overall'])
plt.title('Top 10 Colleges for producing High Performance NBA players')
plt.xlabel('College')
plt.xticks(rotation=90)
plt.ylabel('Overall Score')
plt.show()
```
